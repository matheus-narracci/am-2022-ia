```python
import cv2
import numpy as np
import pytesseract
from matplotlib import pyplot as plt

# FunÃ§Ã£o que mostra a imagem (tons de cinza)
def mostra_imagem(img, title=''):
    plt.figure()
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.xticks([]), plt.yticks([])
    plt.show()

# Carrega a imagem
img = cv2.imread('Cv.png', cv2.IMREAD_GRAYSCALE)
mostra_imagem(img, 'Original')
#Cria o kernel do filtro da mÃ©dia (5x5)
kernel = np.ones((3,3),np.float32)/25
# Filtra a imagem com o kernel criado, mantendo o mesmo tipo da imagem
img_filt = cv2.filter2D(img,-1,kernel)
mostra_imagem(img_filt, 'MÃ©dia 5x5')
```

```python
#Testando qual tipo de filtro fica melhor para leitura do tesseract
img = cv2.imread('Cv.png', cv2.IMREAD_GRAYSCALE)
mostra_imagem(img, 'Original')
median = cv2.medianBlur(img, 1)
mostra_imagem(median, 'Filtro da Mediana 1x1')
```

```python
img_gau = cv2.imread('Cv.png', cv2.IMREAD_GRAYSCALE)
mostra_imagem(img_gau,'Original')
blur = cv2.GaussianBlur(img_gau, (1,1),0)
mostra_imagem(blur,'Filtro gaussiano')
```

```python
#Escolhemos que a suluÃ§Ã£o que melhor se aplica Ã© a mediana -- adicionamos brilho
brilho = 10
imagem_nova = cv2.add(median, brilho)
mostra_imagem(imagem_nova) 
```
